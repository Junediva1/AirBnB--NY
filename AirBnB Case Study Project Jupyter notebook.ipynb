```python
# How many neighborhood groups are available and which shows up the most?
_group=air_bnb.groupby('neighbourhood_group').count()
print(_group)

# 5 neighborhood groups that are available. Brox, Brooklyn, Manhattan, Queens, and Staten Island
```

```python
# Are private rooms the most popular in manhattan?
air_bnb = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan']
room_type_counts = air_bnb['room_type'].value_counts() 
print(room_type_counts)

# Private rooms are not popular in Manhattan, The most popular rooms in Manhattan are Entire home/apt
```

```python
# Which hosts are the busiest and based on their reviews?
host_review = air_bnb.groupby('host_id')['number_of_reviews'].sum()
busiest_id = host_review.idxmax()
busiest_review = host_review.max()
host_info = air_bnb.loc[air_bnb['host_id']== busiest_id,['host_id','host_name']]
print(f"the busiest host is {host_info ['host_name'].iloc[0]},they have {busiest_review} reviews. ")

# Maya is the busiest host who has 2273 reviews.
```

```python
review = air_bnb['number_of_reviews'].idxmax()
print(air_bnb['host_name'].iloc[review],review)
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby(['neighbourhood_group']).mean(numeric_only=True).sort_values('price', ascending=False)
# Manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby(['neighbourhood_group']).sum(numeric_only=True).sort_values('price', ascending=False)
# Manhattan has the highest total price
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_name']).sum(numeric_only=True).sort_values('price', ascending=False).reset_index()[['host_name', 'price']][:5]
# Sonder is the host that has the hightest total price of 82795
# Blueground has 70331
# Micheal has 66895
# David has 65844
# Alex 52563
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
#host_group(['host_name', 'number_of_reviews', 'availability_365'])
listings= air_bnb[(air_bnb.availability_365 == 0)& (air_bnb.number_of_reviews >= 100)]
listings.drop_duplicates('host_id')[['host_name', 'name', 'number_of_reviews']].reset_index(drop=True)
# MaryEllen has no (zero)availability with a review count of 100 
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby(['host_name', 'neighbourhood_group']).sum(numeric_only=True).sort_values('price', ascending=False).reset_index()[0:1]

# Sonder has the highest total of prices in Manhattan
```

```python
# When did Danielle from Queens last receive a review?

dan = (air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] =='Queens')
air_bnb[dan].sort_values('last_review', ascending=False)[:1]

# Danielle received a review on 2019-07-08
```
